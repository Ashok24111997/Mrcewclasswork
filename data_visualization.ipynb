```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
data = pd.DataFrame({
    'category' : np.random.choice(['A', 'B', 'C', 'D'], 100),
    'values' : np.random.randn(100)*10 + 50,
    'X': np.linspace(1,100,100),
    'Y' : np.random.randn(100).cumsum()
})
```

```python
data
```

## Line Plot

used for trend over time or squential data

```python
plt.figure(figsize = (7,4))
plt.plot(data['X'], data['Y'], color = 'blue', linestyle = '-', marker = '+', label = 'Trend')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

## Bar Plot

Used for comparing categories

```python
plt.figure(figsize = (6,4))
sns.barplot(x = 'category', y = 'values', data = data, ci = None, palette = 'Set2')
plt.title('Bar plot example')
plt.show()
```

# Count Plot

want to show frequency of categories

```python
plt.figure(figsize = (6,4))
sns.countplot(x = 'category', data = data, palette = 'pastel')
plt.title('Count Plot Example')
plt.show()
```

# Histogram

shows distribution of numerical data

```python
plt.figure(figsize = (6,4))
plt.hist(data['values'], bins = 10, color = 'skyblue', edgecolor = 'black')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

# Box Plot

shows spread, median and outliers.

```python
plt.figure(figsize = (6,4))
sns.boxplot(x = 'category', y = 'values', data = data, palette = 'coolwarm')
plt.title('Box plot Example')
plt.show()
```

# Scatter plot

used to see relationship between 2 variables

```python
plt.figure(figsize = (6,4))
plt.scatter(data['X'], data['values'], c = 'red', marker = '*', alpha = 0.6)
plt.title('Scatter plot Example')
plt.xlabel('X')
plt.ylabel('Values')
plt.show()
```

# Heatmap

shows correlation or intensity using colors

```python
print(data.corr())
```

```python
plt.figure(figsize = (6,4))
sns.heatmap(data.corr(), annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Heatmap Example')
plt.show()
```

# Pie chart

shows proportions of category

```python
counts = data['category'].value_counts()
plt.figure(figsize = (6,6))
plt.pie(counts, labels = counts.index, autopct = '%1.1f%%', colors = ['skyblue','orange','pink','green'])
plt.title('Pie Chart Example')
plt.show()
```
