```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.random.randint(150,200,50)
data
```

```python
df = data.reshape(-1,1)
df
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)
```

```python
cent =kmeans.cluster_centers_
label = kmeans.labels_
```

```python
print("centroid",cent)
print("label",label)
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(df, np.zeros_like(df),  cmap='viridis', s=100)   #(150,0)
plt.scatter(cent, [0,0,0] , c='red', s=200, marker='X', label='Centroids') #[0,0,0]
plt.title('K-Means Clustering of Student Heights')
plt.xlabel('Height (cm)')
plt.legend()
plt.grid(True)
plt.show()
```
