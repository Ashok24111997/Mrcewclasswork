```python
import  matplotlib.pyplot as plt

exams = [1, 2, 3, 4, 5]
marks = [50, 60, 65, 70, 80]

plt.plot(exams, marks, marker='o', color='blue')
plt.title("Student Marks Trend")
plt.xlabel("Exam Number")
plt.ylabel("Marks")
plt.grid()

plt.show()
```

```python
students = ['A', 'B', 'C', 'D', 'E']
marks = [70, 85, 60, 90, 75]

plt.bar(students, marks, color='green')
plt.title("Students Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
ages = [18, 19, 20, 21, 22, 22, 23, 24, 25, 26, 26, 27]

plt.hist(ages, bins=5, color='orange', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
height = [150, 155, 160, 165, 170, 175]
weight = [50, 55, 60, 65, 70, 75]

plt.scatter(height, weight, color='red')
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")

plt.show()
```

```python
activities = ['Study', 'Sleep', 'Play', 'Others']
hours = [6, 8, 4, 6]

plt.pie(hours, labels=activities, autopct='%1.1f%%')
plt.title("Daily Activities")

plt.show()
```
